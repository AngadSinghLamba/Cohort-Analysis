# #  Unveiling the Top 5 Years Through Global Sales 🌐📊
#\
Revel in the dynamic spectrum of gaming preferences over the years, where the perennial dominance of Action harmonizes with diverse genres, crafting a rich and intricate tapestry within the gaming landscape. 🎮✨

```python
df['Genre'].value_counts()  #Note as we can see that the action genre has been made the most while most of the gamers preferred it.
plt.figure(figsize=(15, 10))
sns.countplot(x="Genre", data=df, order=df['Genre'].value_counts().index)
plt.xticks(rotation='vertical')
plt.title('Games Most Released Genre Wise', fontsize=40)
plt.ylabel('Count', fontsize=25)
plt.xlabel('Genre', fontsize=25)
plt.show()
```

```python
# Finding the top 5 years based on the number of game releases
top5_years = df['Year'].value_counts().sort_values(ascending=False).head(5).index

# Filtering the DataFrame for the top 5 years
df_top5_years = df[df['Year'].isin(top5_years)]

# Grouping by Year and Genre, then counting occurrences
yearly_genre_counts = df_top5_years.groupby(['Year', 'Genre']).size().reset_index(name='Counts')

# Finding the top 5 genres for each of the top 5 years
top5_genres_per_year = yearly_genre_counts.groupby('Year').apply(lambda x: x.nlargest(5, 'Counts'))

# Displaying top 5 genres numerically
print("Top 5 Genres for Top 5 Years:")
print(top5_genres_per_year[['Year', 'Genre', 'Counts']])

# Plotting the results with 'viridis' color palette
plt.figure(figsize=(15, 10))
sns.barplot(x='Year', y='Counts', hue='Genre', data=top5_genres_per_year, palette='viridis')
plt.title('Top 5 Genres for Top 5 Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.legend(title='Genre', title_fontsize='15')
plt.show()
```

# 🎮 Top 5 Games Insight:
Across the top 5 gaming years, certain titles stood out with consistent popularity. Notable mentions include:

- 🍲 Ratatouille in 2007, showcasing enduring appeal.
- 🚗 Need For Speed: Undercover and 🎮 FIFA Soccer 10 in 2008 and 2009, highlighting racing and sports genres' ongoing influence.
- 🧙‍♂️ Harry Potter and the Half-Blood Prince, LEGO Harry Potter: Years 1-4, and FIFA 12 in 2009 and 2010, underlining the enduring magic of fantasy and sports genres.

These games not only reflect diverse gaming tastes but also demonstrate the staying power of certain franchises over the years. 🕹️✨

```python
yearly_game_counts = df_top5_years.groupby(['Year', 'Name']).size().reset_index(name='Counts')


top5_games_per_year = yearly_game_counts.groupby('Year').apply(lambda x: x.nlargest(5, 'Counts'))

# Displaying top 5 games numerically
print("Top 5 Games for Top 5 Years:")
print(top5_games_per_year[['Year', 'Name', 'Counts']])

# Plotting the results with a custom contrasting color palette
colors = ['#1f78b4', '#33a02c', '#e31a1c', '#ff7f00', '#6a3d9a']
plt.figure(figsize=(15, 10))
sns.barplot(x='Year', y='Counts', hue='Name', data=top5_games_per_year, palette=colors)
plt.title('Top 5 Games for Top 5 Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.legend(title='Game', title_fontsize='15')
plt.show()
```

```python
top_publishers = df['Publisher'].value_counts().head(5)

# Displaying top 5 publishers
print("Top 5 Publishers Based on All Years:")
print(top_publishers)
```

```python
# Grouping by Publisher and summing up global sales
top_publishers_global_sales = df.groupby('Publisher')['Global_Sales'].sum().nlargest(5)

# Displaying top 5 publishers based on global sales
print("Top 5 Publishers Based on Global Sales:")
print(top_publishers_global_sales)
```

```python


# Calculating average sales per game for each publisher
avg_sales_per_game_publisher = df.groupby(['Publisher', 'Name'])['Global_Sales'].mean().groupby('Publisher').mean()

# Finding the top 5 publishers based on average sales per game
top_publishers_avg_sales = avg_sales_per_game_publisher.nlargest(5)

# Displaying top 5 publishers based on average sales per game
print("Top 5 Publishers Based on Average Sales Per Game:")
print(top_publishers_avg_sales)
```

**As we can see in the above diagram that 2008 and 2009 has been the highest sale worldwide.**

```python
# Assuming df is your DataFrame
plt.figure(figsize=(14, 8))

# Box plot
sns.boxplot(x='Genre', y='Global_Sales', data=df, palette='viridis')
plt.title("Global Sales by Genre")
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Assuming your DataFrame is named 'df' and contains 'Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', and 'Year' columns

# Scatter plot for Global Sales vs. NA Sales, EU Sales, JP Sales
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x='NA_Sales', y='Global_Sales', data=df)
plt.title('Global Sales vs. NA Sales')

plt.subplot(1, 3, 2)
sns.scatterplot(x='EU_Sales', y='Global_Sales', data=df)
plt.title('Global Sales vs. EU Sales')

plt.subplot(1, 3, 3)
sns.scatterplot(x='JP_Sales', y='Global_Sales', data=df)
plt.title('Global Sales vs. JP Sales')

plt.tight_layout()
plt.show()

# Line plot for Year and Global Sales
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Global_Sales', data=df)
plt.title('Year and Global Sales Trends')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show()

# Correlation matrix heatmap
correlation_matrix = df[['Global_Sales', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Year']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

📊 Insights on Video Game Sales Data:

🌐 Global Sales vs. Other Regions:

The global sales show a strong negative correlation with North American (NA) and European (EU) sales, indicating that as one region's sales increase, the global sales decrease.
However, the correlation with Japanese (JP) sales is less significant, suggesting a more complex relationship.
🕰️ Year and Sales Trends:

The positive correlation between the year and global sales is relatively weak, implying that the year of release may not have a substantial impact on overall sales trends.
🌍 Regional Sales Relationships:

Notably, strong positive correlations between NA and EU sales, as well as NA and JP sales, indicate similar preferences in these regions.
The EU and JP sales also exhibit a positive correlation, suggesting shared gaming interests.
🛍️ Other Sales Contributions:

The positive correlations between global sales and other sales (outside major regions) highlight the importance of diverse market contributions to overall game sales.
📈 Overall Implications:

To optimize global sales, understanding and tailoring marketing strategies for specific regional preferences is crucial. The complex interplay between regions necessitates a nuanced approach for successful game releases.
